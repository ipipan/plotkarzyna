```python
from pathlib import Path

import coref_ds
from coref_ds.text import Text
from coref_ds.mmax.mmax_doc import MmaxDoc
from coref_ds.tei.tei_doc import TEIDocument

from plotkarzyna.config import *
```

```python
mmax_paths = list(Path(local_config['NKJP_MMAX_ROOT']).glob('**/*.mmax'))
tei_train_paths = list(
    (Path(local_config['PCC_ROOT']) / 'train').iterdir()
    )
```

```python
mmax_docs = []
for p in mmax_paths:
    try: 
        m = MmaxDoc.from_file(p)
        mmax_docs.append(m)
    except Exception as e:
        print(e)
```

```python
tei_docs = []
for p in tei_train_paths:
    try: 
        m = TEIDocument(p)
        tei_docs.append(m)
    except Exception as e:
        print(e)
```

```python
texts = []
for m in mmax_docs:
    texts.append(m.text)
```

```python
texts = []
for m in tei_docs:
    texts.append(m.text)
```
